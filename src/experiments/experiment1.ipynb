# Experiment 1
Preliminary experiment conducted to see if any pytorch transformations will help us better see the difference between a weak positive that is false and weak positive that is actually a positive.

```python
# Import section
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
IMG_PATH="./samples"
```

```python
def apply_clahe(img_file, title=''):
    image = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    clahe_image = clahe.apply(image)

    # Display the results
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    axes[1].imshow(clahe_image, cmap='gray')
    axes[1].set_title("CLAHE Enhanced Image")
    axes[1].axis("off")
    
    fig.suptitle(title)

    plt.show()
```

```python
weak_false_positive = os.path.join(IMG_PATH, 'weak_positive_false.jpg')
weak_true_positive = os.path.join(IMG_PATH, 'weak_positive_true.JPG')
```

```python
apply_clahe(weak_false_positive, title="Weak False Positive")
```

```python
apply_clahe(weak_true_positive, title="Weak True Positive")
```
